```python
import os
import csv
import pandas as pd
```

```python
csv_schools = os.path.join("Resources", "schools_complete.csv")
csv_students = os.path.join("Resources", "students_complete.csv")
```

```python
schools_data = pd.read_csv(csv_schools)
students_data = pd.read_csv(csv_students)
```

```python
education_data_df = pd.merge(students_data, schools_data, how="left", on=["school_name", "school_name"])
education_data_df.head()                                                                           
```

District Summary

```python
unique_schools = len(education_data_df["school_name"].unique())
print(unique_schools)
```

```python
unique_students = education_data_df["Student ID"].count()
print(unique_students)
```

```python
total_budget = education_data_df["budget"].unique()
total_budget = total_budget.sum()
print(total_budget)
```

```python
average_math_score = education_data_df["math_score"].mean()
print(average_math_score)
```

```python
average_reading_score = education_data_df["reading_score"].mean()
print(average_reading_score)
```

```python
passing_math_count = education_data_df[(education_data_df["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(unique_students) * 100
print(passing_math_percentage)
```

```python
passing_reading_count = education_data_df[(education_data_df["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(unique_students) * 100
print(passing_reading_percentage)
```

```python
passing_math_reading_count = education_data_df[
    (education_data_df["math_score"] >= 70) & (education_data_df["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(unique_students) * 100
overall_passing_rate
```

```python
district_summary = {
    "School Count":unique_schools,
    "Total Students":unique_students,
    "Total Budget":total_budget,
    "Average Math Score":average_math_score,
    "Average Reading Score":average_reading_score,
    "% Passing Math":passing_math_percentage,
    "% Passing Reading":passing_reading_percentage,
    "% Overall Passing":overall_passing_rate
}

district_summary = pd.DataFrame([district_summary])
# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
# Display the DataFrame
district_summary
```

School Summary

```python
school_type = education_data_df.groupby(['school_name'])['type'].first()
school_type
```

```python
per_school_counts = education_data_df.groupby(['school_name'])['Student ID'].count()
per_school_counts
```

```python
per_school_budget = education_data_df.groupby(['school_name'])['budget'].mean()
per_school_budget
```

```python
per_school_capita = per_school_budget / education_data_df.groupby(['school_name'])['School ID'].count()
per_school_capita
```

```python
per_school_reading = education_data_df.groupby(['school_name'])['reading_score'].sum() / education_data_df.groupby(['school_name'])['School ID'].count()
per_school_reading
```

```python
per_school_math = education_data_df.groupby(['school_name'])['math_score'].sum() / education_data_df.groupby(['school_name'])['School ID'].count()
per_school_math                                        
```

```python
passing_math_all = education_data_df.loc[education_data_df['math_score'] > 70 , ['Student ID', 'school_name', 'math_score']]
passing_math_all.groupby(['school_name'])['Student ID'].count()
```

```python
passing_read_all = education_data_df.loc[education_data_df['reading_score'] > 70 , ['Student ID', 'school_name', 'reading_score']]
passing_read_all.groupby(['school_name'])['Student ID'].count()
```

```python
passing_both_all = education_data_df[(education_data_df["reading_score"] >= 70) & (education_data_df["math_score"] >= 70)]
passing_both_all = passing_both_all.groupby(["school_name"]).size()
print(passing_both_all)
```

```python
per_school_passing_math = passing_math_all.groupby(['school_name'])['Student ID'].count() / education_data_df.groupby(['school_name'])['School ID'].count() * 100
per_school_passing_reading = passing_read_all.groupby(['school_name'])['Student ID'].count() / education_data_df.groupby(['school_name'])['School ID'].count() * 100
overall_passing_rate = passing_both_all / education_data_df.groupby(['school_name'])['School ID'].count() * 100
```

```python
per_school_summary = pd.DataFrame({
    "School Type":school_type,
    "Total Students":per_school_counts,
    "Total School Budget":per_school_budget,
    "Per Student Budget":per_school_capita,
    "Average Math Score":per_school_math,
    "Average Reading Score":per_school_reading,
    "% Passing Math":per_school_passing_math,
    "% Passing Reading":per_school_passing_reading,
    "% Overall Passing":overall_passing_rate
})
```

```python
per_school_summary_format= per_school_summary.copy()
```

```python
# Formatting
per_school_summary_format["Total Students"] = per_school_summary_format["Total Students"].map("{:,}".format)
per_school_summary_format["Total School Budget"] = per_school_summary_format["Total School Budget"].map("${:,.2f}".format)
per_school_summary_format["Per Student Budget"] = per_school_summary_format["Per Student Budget"].map("${:,.2f}".format)
per_school_summary_format["% Passing Math"] = per_school_summary_format["% Passing Math"].map("%{:,.2f}".format)
per_school_summary_format["% Passing Reading"] = per_school_summary_format["% Passing Reading"].map("%{:,.2f}".format)
per_school_summary_format["% Overall Passing"] = per_school_summary_format["% Overall Passing"].map("%{:,.2f}".format)
per_school_summary_format
```

High Performing Schools (by % Overall Passing)

```python
top_schools = per_school_summary.sort_values(by=['% Overall Passing'], ascending=False)
top_schools.head()
```

Poor Performing Schools (by % Overall Passing)

```python
bottom_schools = per_school_summary.sort_values(by=['% Overall Passing'], ascending=True)
bottom_schools.head()
```

Math Scores by Grade

```python
ninth_graders = education_data_df[(education_data_df["grade"] == "9th")]
tenth_graders = education_data_df[(education_data_df["grade"] == "10th")]
eleventh_graders = education_data_df[(education_data_df["grade"] == "11th")]
twelfth_graders = education_data_df[(education_data_df["grade"] == "12th")]
```

```python
ninth_grade_math_scores = ninth_graders.groupby(['school_name'])['math_score'].mean()
tenth_grade_math_scores = tenth_graders.groupby(['school_name'])['math_score'].mean()
eleventh_grade_math_scores = eleventh_graders.groupby(['school_name'])['math_score'].mean()
twelfth_grade_math_scores = twelfth_graders.groupby(['school_name'])['math_score'].mean()
```

```python
df1 = pd.merge(ninth_grade_math_scores, tenth_grade_math_scores, how="left", on=["school_name"])
df2 = pd.merge(eleventh_grade_math_scores, twelfth_grade_math_scores, how="left", on=["school_name"])
math_scores_by_grade = pd.merge(df1, df2, how="left", on=["school_name"])                                
```

```python
math_scores_by_grade.index.name = None
math_scores_by_grade.rename(columns={"math_score_x_x": "9th", "math_score_y_x": "10th", "math_score_x_y": "11th", "math_score_y_y": "12h"})
```

Reading Scores by Grade

```python
ninth_graders = education_data_df[(education_data_df["grade"] == "9th")]
tenth_graders = education_data_df[(education_data_df["grade"] == "10th")]
eleventh_graders = education_data_df[(education_data_df["grade"] == "11th")]
twelfth_graders = education_data_df[(education_data_df["grade"] == "12th")]
```

```python
ninth_grade_reading_scores = ninth_graders.groupby(['school_name'])['reading_score'].mean()
tenth_grade_reading_scores = tenth_graders.groupby(['school_name'])['reading_score'].mean()
eleventh_grade_reading_scores = eleventh_graders.groupby(['school_name'])['reading_score'].mean()
twelfth_grade_reading_scores = twelfth_graders.groupby(['school_name'])['reading_score'].mean()
```

```python
df1_1 = pd.merge(ninth_grade_reading_scores, tenth_grade_reading_scores, how="left", on=["school_name"])
df2_2 = pd.merge(eleventh_grade_reading_scores, twelfth_grade_reading_scores, how="left", on=["school_name"])
reading_scores_by_grade = pd.merge(df1_1, df2_2, how="left", on=["school_name"])      
```

```python
reading_scores_by_grade.index.name = None
reading_scores_by_grade.rename(columns={"reading_score_x_x": "9th", "reading_score_y_x": "10th", "reading_score_x_y": "11th", "reading_score_y_y": "12h"})
```

Scores by School Spending

```python
school_spending_df = per_school_summary.copy()
```

```python
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
school_spending_df["Spending Ranges"] = pd.cut(school_spending_df['Per Student Budget'], spending_bins, labels=labels)
school_spending_df.head()
```

```python
spending_math_scores = school_spending_df.groupby(["Spending Ranges"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges"])["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges"])["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges"])["% Overall Passing"].mean()
```

```python
spending_summary_df = pd.DataFrame({
    "Average Math Score":spending_math_scores,
    "Average Reading Score":spending_reading_scores,
    "% Passing Math":spending_passing_math,
    "% Passing Reading":spending_passing_reading,
    "% Overall Passing":overall_passing_spending,
})
spending_summary_df  
```
